```python
import tensorflow as tf
from tensorflow.keras import models,layers
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D,Input, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt
```

```python
IMAGE_SIZE = 256
BATCH_SIZE = 32
CHANNELS = 3
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "Train",
    shuffle=True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE,
)
```

```python
valid_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "Valid",
    shuffle=True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE,
)
```

```python
class_name = train_ds.class_names
class_name
```

```python
len(train_ds)
```

```python
plt.figure(figsize=(12,12))
for image_batch, label_batch in train_ds.take(1):
    for i in range(12):
        ax = plt.subplot(3,4,i+1)
        plt.imshow(image_batch[i].numpy().astype("uint8"))
        plt.title(class_name[label_batch[i]])
    print(image_batch[i].shape)
```

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
valid_ds = valid_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(IMAGE_SIZE,IMAGE_SIZE),
    layers.experimental.preprocessing.Rescaling(1.0/255)
    ])
```

```python
data_augmentation = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
    layers.experimental.preprocessing.RandomRotation(0.2),
])
```

```python
train_ds = train_ds.map(
    lambda x, y: (data_augmentation(x, training=True), y)
).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
input_shape = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
num_classes = 4

# Create the CNN model
input_cnn = Input(shape=input_shape, batch_size=BATCH_SIZE)
conv1 = Conv2D(32, kernel_size=(3, 3), activation='relu')(input_cnn)
pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
conv2 = Conv2D(64, kernel_size=(3, 3), activation='relu')(pool1)
pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)
flatten = Flatten()(pool2)

# Create the MLP model
input_mlp = Input(shape=(flatten.shape[1],))
dense1 = Dense(128, activation='relu')(input_mlp)
dropout = Dropout(0.5)(dense1)
dense2 = Dense(64, activation='relu')(dropout)

# Combine the CNN and MLP models
combined = tf.keras.layers.concatenate([flatten, dense2])
output = Dense(num_classes, activation='softmax')(combined)

# Create the hybrid model
model = Model(inputs=[input_cnn, input_mlp], outputs=output)
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```

```python
model.fit(
    [x for x in train_ds],  # Pass both CNN and MLP inputs
    epochs=10,
    validation_data=valid_ds
)
```
